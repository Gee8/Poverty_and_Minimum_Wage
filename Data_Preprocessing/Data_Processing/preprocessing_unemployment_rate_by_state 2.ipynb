```python
def process_unemployment(path, filename):
    # path to individual data file
    path_filename = path + filename
    # read csv into df
    unemp_df = pd.read_csv(path_filename)
    unemp_df.columns=["DATE", "unemployment_rate"]
    # split filename to get the name of the state and add to the column state
    state_name = filename.split('_')[0]
    unemp_df ['state'] = state_name
    print(path_filename)
    # renaming the column name to unemployment_rate
    #unemp_df .rename(index = {1: 'unemployment_rate'})
    print(unemp_df.head(5))
    # formatting the year column to only display year
    unemp_df ['year']= pd.to_datetime(unemp_df['DATE'], errors='coerce').dt.year
    unemp_df .drop(columns = ['DATE'], inplace = True)
    
    return unemp_df 
```

```python
import os
import sys

unemployment_df = process_unemployment("", "National_Unemployment_rate_1976_2020.csv")
path = ('Unemployment/')

for filename in os.listdir(path):
    if filename[0] != ".":
        
        state_df = process_unemployment(path, filename)
        unemployment_df = pd.concat([unemployment_df, state_df], axis = 0, ignore_index = True)
    
    
unemployment_df.head(60)
```
