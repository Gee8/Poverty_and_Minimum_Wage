## Generating queries to populate lowest/highest poverty rate tables

```python
import pandas as pd
```

```python
df = pd.read_csv('../Data_Preprocessing/Raw_Data/dataset_na_dropped.csv')
df.head()
```

```python
# Generating code to insert rows into lowest_poverty_rates table
for state in df['state'].unique():
    print(f"-- Reading in lowest poverty rate data for {state}\nINSERT INTO lowest_poverty_rates\nSELECT * FROM economic_features AS e\nWHERE e.state = '{state}'\nORDER BY e.poverty_rate\nFETCH FIRST ROW ONLY;\n")
```

```python
# Generating code to insert rows into highest_poverty_rates table
for state in df['state'].unique():
    print(f"-- Reading in highest poverty rate data for {state}\nINSERT INTO highest_poverty_rates\nSELECT * FROM economic_features AS e\nWHERE e.state = '{state}'\nORDER BY e.poverty_rate DESC\nFETCH FIRST ROW ONLY;\n")
```
